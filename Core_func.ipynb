```python
#for numpy
import numpy as np
```

```python
#for plotting purpose
import matplotlib.pyplot as plt
```

```python
#for scikit learn library
from sklearn import tree
```

```python
#to deal with .dot extension files
import pydotplus
```

```python
#to render tree image
import graphviz
```

```python
#to remove a file
import os
```

```python
#visualize dataset
def visualize_dataset(x,y,vec):
    on=(y==1).ravel()
    off=(y==0).ravel()
    plt.scatter(x[on,0],x[on,1],c='b',marker='+')
    plt.scatter(x[off,0],x[off,1],c='r',marker='o')
    plt.xlabel(vec[0])
    plt.ylabel(vec[1])
    plt.legend(["y=1","y=0"])
```

```python
#visualize tree
def visualize_tree(clf,filename,vec):
    filename_dot=filename+".dot"
    filename_png=filename+".png"
    tree.export_graphviz(clf,out_file=filename_dot,rounded=True,filled=True,feature_names=vec)
    graph=pydotplus.graph_from_dot_file(filename_dot)
    graph.write_png(filename_png)
    os.remove(filename_dot)
    image=plt.imread(filename_png)
    plt.figure(figsize=(15,15))
    plt.imshow(image)
```

```python
#to find error in a given data set using parameters of tree trained by training dataset
def error_in_set(x,y,clf):
    right=0
    m=len(y)
    for i in range(0,m):
        if clf.predict([x[i,:]])==y[i]:
            right+=1
    error=((m-right)/m)*100
    return error
```

```python
#location of ex2data1.txt file
datafileloc1="C:/Users/Ritik/Desktop/Projects_Flash/Decision Tree/ex2data1.txt"
```

```python
#loading csv file contents in the program
data1=pd.read_csv(datafileloc1,header=None,names=["Score_1","Score_2","Admission Decision"])
```

```python
print(data1)
```

```python
#extracting input and output examples from training set
x=np.array(data1.iloc[:,0:2])
y=np.array(data1.iloc[:,2])
```

```python
#knowing about dimensions of those extracted data points
print(x.shape)
print(y.shape)
```

```python
#visualizing dataset
visualize_dataset(x,y,["Score_1","Score_2"])
```

```python
#training tree with scikit learn library
clf=tree.DecisionTreeClassifier()
clf=clf.fit(x,y)
```

```python
#visualizing tree
visualize_tree(clf,"Treespy",["Score_1","Score_2"])
```

```python
#checking error in training set
#observe that it overfits
error_in_set(x,y,clf)
```
